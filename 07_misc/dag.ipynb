# Directed Acyclic Graphs (DAGs)

A **DAG** is a graph where:

- **Nodes** = variables
- **Directed edges (→)** = causal effects
- **Acyclic** = you can't start at one node and come back to it by following arrows

Example:

Smoking → Lung Cancer

Suppose we want to answer: Does studying more hours improve exam scores?

- X = hours studied
- Y = exam score

Naively:

X → Y

### Confounding

Now consider **motivation**:

- Motivated students:
  - study more
  - perform better

Let:
- Z = motivation

Now the causal structure is:

Z → X  
Z → Y  

Motivation affects **both** the cause and the outcome.

This is called **confounding**, where Z is the confounder.

So X and Y are **correlated** but there might not be **causation**

### Controlling for a Collider

Now consider this structure:

X → Z ← Y

Example:
- X = studying
- Y = intelligence
- Z = being admitted to an elite school

Both X and Y affect Z.

Z is a **collider**.

We do NOT want to condition on a collider Z because that makes X and Y look related even though they might not be
