# Autoencoders

An autoencoder is a neural network trained to reconstruct its input:

Input $\rightarrow$ Encode $\rightarrow$ Latent vector

Latent vector $\rightarrow$ Decode $\rightarrow$ Reconstruction (Estimate of input)

Used in:

- Dimensionality reduction

- Anomaly detection

- Image compression

- Noise removal

```python
import torch
import torch.nn as nn
```

Example:

Autoencoder for 28 x 28 images

```python
class Autoencoder(nn.Module):
    def __init__(self):
        super().__init__()
        
        # 28*28 -> 32
        self.encoder = nn.Sequential(
            nn.Linear(28*28, 128),
            nn.ReLU(),
            nn.Linear(128, 32)
        )
        
        # 32 -> 28*28
        self.decoder = nn.Sequential(
            nn.Linear(32, 128),
            nn.ReLU(),
            nn.Linear(128, 28*28),
            nn.Sigmoid()
        )
        
    def forward(self, x):
        x = x.view(x.size(0), -1)
        z = self.encoder(x)
        out = self.decoder(z)
        return out
```

Here z is the encoded representation.
